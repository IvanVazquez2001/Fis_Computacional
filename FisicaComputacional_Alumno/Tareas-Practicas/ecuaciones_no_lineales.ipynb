## biseccion

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (x**3)+7

x=np.linspace(-5,5,1000)
y=[]
for i in x:
    y.append(f(i))

plt.plot(x,y,color='green')
plt.hlines(y=0,xmin=-5,xmax=5,color='black',linestyle='--')
plt.grid(True)
```
