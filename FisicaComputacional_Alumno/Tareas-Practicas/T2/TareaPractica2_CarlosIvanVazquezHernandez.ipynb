# La constante de Madelung

En física condensada la constante de Mandelug es el potencial electrico total que siente un átomo en un sólido y viene dado por:
\begin{equation}
    V_{total}=\sum^{L}_{i,j,k=-L\\; i,j,k\not =0} V(i,j,k)=\frac{e}{4\pi\epsilon_0a}M
\end{equation}
Donde el valor de $V(i,j,k)$ es:
\begin{equation}
    V(i,j,k)=\pm\frac{e}{4\pi\epsilon_0a\sqrt{i^{2}+j^{2}+k^{2}}}
\end{equation}
Si $i+j+k$ es par entonces el signo es positivo, si es impar el signo es negativo. La constante se obtiene cuando L tiene a infinito, asi que se calcula la constante con el valor de L mas grande sin dejar que el orgrama se ejecute en mas de un minuto.

```python
from math import *
def Constante_de_mandelug(L):
    m=[]
    ### como tenemos un suma en tres dimensiones hacemos tres ciclos for donde cada ciclo 
    ### representa la suma sobre una dimensión
    for i in range(-L,L+1):
        for j in range(-L,L+1):
            for k in range(-L,L+1):
                ### si i+j+k es par hacemos el signo positivo
                if (i+j+k)%2==0:
                    signo=1
                else:
                    signo=-1
                ### La suma no esta definida para i=j=k=0 por lo que ponemos un continue para saltar
                ### ete valor sin salir del loop.
                if i==0 or j==0 or k==0:
                    continue
                else:
                    m.append(signo/sqrt(i**2+j**2+k**2))
    return sum(m)
```

# Coeficientes Binomiales

(a) Escribe una función llamada binomial(n,k) que calcule el coeficiente binoimal para un n y k dados y que regrese el valor en número entero.
La fórmula del coeficiente binomial es:
\begin{equation}
    \binom{n}{k}=\frac{n!}{k!(n-k)!}
\end{equation}

```python
def binomial(n,k):
    nf=1
    kf=1
    nkf=1
    if k==0:
        comb=1
    else:
        #### con los ciclos for obtenemos los factoriales de n,k y (n-k) respectivamente.
        for j in range(1,n+1):
            nf=nf*j
        for m in range(1,k+1):
            kf=kf*m
        for i in range(1,(n-k)+1):
            nkf=nkf*i
        comb=int(nf/(kf*nkf))
    return (comb)
```

# Numeros primos

Escribe un programa que nos diga si un número entre 3 y 10,000 es primo, las condiciones para que un número NO sea primo son: Si encontramos un factor primo que sea menor a $\sqrt{n}$

```python
from math import *
def Numeros_primos(n):
    primos=[2]
    for i in range(3,n+1):
        ### Para cada número de la lista le stamos diciendo que si es primo.
        primo=True
        for j in range(2,i):
            ### Estas son las dos condiciones que hacen que un número no sea primo, por lo que si se cumplen, 
            ### cambiamos el valor de true a false
            if i%j == 0 and j**2<=i:
                primo=False
            ### Si no se cumplen entonces nos seguimos al siguiente valor dentro del loop y dejamos primo=True
            else:
                continue
        ### aqui se agrega el numero a la lista
        if primo==True:
            primos.append(i)
                
            
            
    return primos
            
```

```python
Numeros_primos(10000)
```

# Recursividad

(a) Usando la recursividad de una función, encuentra el valor de $C_{100}$ el cual es un número de Catalan.
\begin{equation}
    f(x)= \left\{ \begin{array}{lcc} 1 & si & n=0 \\ \\ \frac{4n-2}{n+1}C_{n-1} & si & n\geq 1\\ \end{array} \right.
\end{equation}

```python
#### INCISO A
def Catalan(n):
    if n==0:
        cn=1
    else:
        cn = ((4*n-2)/(n+1))*Catalan(n-1)    
    
    return int(cn)
```

```python
print(Catalan(100))
```
